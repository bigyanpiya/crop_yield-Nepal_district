```python
# Renaming the column.  
dataframe_yield.rename( columns={"Value": "hg/ha_yield"}, inplace=True)  
dataframe_yield.head()  
```

(hectogram per hectare (Hg/Ha))

```python
# dropping the unwanted columns.  
dataframe_yield = dataframe_yield.drop(['Year Code','Element Code', 'Element','Area Code (M49)','Domain Code', 'Domain','Unit','Item Code (CPC)'], axis=1)  
 
```

```python
dataframe_yield.tail()
```

```python
# Group by Item to find the total yield for each food item
total_yield_per_item = dataframe_yield.groupby('Item')['hg/ha_yield'].sum().sort_values(ascending=False)

print(total_yield_per_item)
```

```python
# Group by Year to find the total yield for each year
total_yield_per_year = dataframe_yield.groupby('Year')['hg/ha_yield'].sum().sort_values(ascending=False)

print(total_yield_per_year)
```

```python
# Plot top yielding food items
plt.figure(figsize=(12, 6))
sns.barplot(x=total_yield_per_item.index, y=total_yield_per_item.values)
plt.xticks(rotation=90)
plt.title('Total Yield per Food Item')
plt.xlabel('Food Item')
plt.ylabel('Total Yield (hg/ha)')
plt.show()
```

```python
# Function to get top yielding crops for a particular year
def top_yielding_crops(year, n=10):
    year_data = dataframe_yield[dataframe_yield['Year'] == year]
    top_crops = year_data.groupby('Item')['hg/ha_yield'].sum().sort_values(ascending=False).head(n)
    return top_crops

# Example for the year 2020
print(top_yielding_crops(2019))
```

```python
# Example for the year 2020
top_crops_2020 = top_yielding_crops(2020)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_crops_2020.index, y=top_crops_2020.values)
plt.xticks(rotation=90)
plt.title('Top Yielding Crops in 2020')
plt.xlabel('Crop')
plt.ylabel('Total Yield (hg/ha)')
plt.show()
```

```python
# Filter the dataset for wheat and maize
df_filtered = dataframe_yield[dataframe_yield['Item'].isin(['Apples', 'Maizea (corn)'])]
```

```python
# Group by Year and Item to get the total yield for wheat and maize each year
yield_per_year = df_filtered.groupby(['Year', 'Item'])['hg/ha_yield'].sum().reset_index()
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")

# Create a line plot to compare yields
plt.figure(figsize=(14, 8))
sns.lineplot(data=yield_per_year, x='Year', y='hg/ha_yield', hue='Item', marker='o')

# Adding title and labels
plt.title('Comparison of Wheat and Maize Yields Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Yield (hg/ha)')
plt.legend(title='Item')
plt.grid(True)
plt.show()
```

```python
# Create a bar plot to compare yields
plt.figure(figsize=(14, 8))
sns.barplot(data=yield_per_year, x='Year', y='hg/ha_yield', hue='Item')

# Adding title and labels
plt.title('Comparison of Wheat and Maize Yields by Year')
plt.xlabel('Year')
plt.ylabel('Total Yield (hg/ha)')
plt.legend(title='Item')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
correlation_data = dataframe_main.select_dtypes(include=[np.number]).corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_data, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlation_data, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
from sklearn.preprocessing import OneHotEncoder  
  
dataframe_main_onehot = pd.get_dummies(dataframe_main, columns=['Area',"Item"], prefix = ['Country',"Item"])  
features=dataframe_main_onehot.loc[:, dataframe_main_onehot.columns != 'hg/ha_yield']  
label=dataframe_main['hg/ha_yield']  
features.head()  
```

```python
#Dropping the year column  
features= features.drop(['Year'], axis=1)  
  
features.info()  
```

```python
from sklearn.preprocessing import MinMaxScaler  
scaler=MinMaxScaler()  
features=scaler.fit_transform(features)  
```

```python
features  
```

```python
from sklearn.model_selection import train_test_split  
train_data, test_data, train_labels, test_labels = train_test_split(features, label, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import r2_score  
def compare_models(model):  
    model_name = model.__class__.__name__  
    fit=model.fit(train_data,train_labels)  
    y_pred=fit.predict(test_data)  
    r2=r2_score(test_labels,y_pred)  
    return([model_name,r2])  
  
from sklearn.ensemble import RandomForestRegressor  
from sklearn.ensemble import GradientBoostingRegressor  
from sklearn import svm  
from sklearn.tree import DecisionTreeRegressor  
  
models = [  
    GradientBoostingRegressor(n_estimators=200, max_depth=3, random_state=0),  
     RandomForestRegressor(n_estimators=200, max_depth=3, random_state=0),  
    svm.SVR(),  
   DecisionTreeRegressor()  
]  
  
model_train=list(map(compare_models,models))  
  
print(*model_train, sep = "\n") 
```

```python
import matplotlib.pyplot as plt

# Define a function to plot actual vs predicted values
def plot_actual_vs_predicted(model, model_name):
    fit = model.fit(train_data, train_labels)
    y_pred = fit.predict(test_data)
    plt.figure(figsize=(8, 6))
    plt.scatter(test_labels, y_pred, alpha=0.5)
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title(f"Actual vs Predicted ({model_name})")
    plt.show()

# Loop through models, fit them, predict, and plot actual vs predicted values
for model, model_name in zip(models, [model.__class__.__name__ for model in models]):
    plot_actual_vs_predicted(model, model_name)
```
